### Augment data with different columns

#### 1. Adding month column:

```python
'04/19/19 08:46'.split('/')[0]
```

```python
def month(x):
    return x.split('/')[0]
```

```python
all_data['Month'] = all_data['Order Date'].apply(month)
```

```python
all_data
```

```python
all_data['Month'] = all_data['Month'].astype(int)
```

```python
all_data['Month'].unique()
```

```python
filter = all_data['Month']=='Order Date'

all_data[filter]
```

```python
all_data = all_data[~filter]
all_data
```

```python
all_data['Month'] = all_data['Month'].astype(int)
```

```python
all_data.dtypes
```

```python
all_data['Quantity Ordered'] = all_data['Quantity Ordered'].astype(int)
```

```python
all_data['Price Each'] = all_data['Price Each'].astype(float)
```

```python
all_data.dtypes
```

#### 2. Adding sales column:

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
```

```python
all_data.head()
```

####  3. Adding city column:

```python
'917 1st St, Dallas, TX 75001'.split(',')[1]
```

```python
'917 1st St, Dallas, TX 75001'.split(',')[2].split(' ')[1]
```

```python
def city(address):
    return address.split(',')[1]

def state(address):
    return address.split(',')[2].split(' ')[1]

```

```python
all_data['City'] = all_data['Purchase Address'].apply(lambda x: city(x)+' '+'('+state(x))+')'
```

```python
df['Grouped Product'] = df.groupby('Order ID')['Product'].transform(lambda x: ', '.join(x))
```

```python
df2 = df[['Order ID','Grouped Product']].drop_duplicates()
df2
```

```python
df2['Grouped Product'].value_counts()
```

```python
df2['Grouped Product'].value_counts()[0:5]
```

```python
grouped_product = df2['Grouped Product'].value_counts()[0:10].index

frequency = df2['Grouped Product'].value_counts()[0:10]

# plt.pie(grouped_product,frequency)
# plt.xticks(grouped_product, rotation='vertical',size=10)
# plt.xlabel('Grouped Product')
# plt.ylabel('Frequency')
# plt.show()

plt.pie(frequency,labels=grouped_product,autopct='%0.2f%%', radius=2)
plt.show()
```
